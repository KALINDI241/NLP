```python
df = open("Gudgau.json",'r')
db=json.load(df)
data=""
for i in range(0,len(db)):
        #print(db[i]['Comment'])
        data=data+db[i]['Comment']
print(data)
```

```python
def N_gram(n,data):
  tokens=re.split("\\s+",data)
  ngrams=[]
  for i in range(len(tokens)-n+1):
    temp=[tokens[j] for j in range (i,i+n)]
    ngrams.append(" ".join(temp))
  return ngrams
```
